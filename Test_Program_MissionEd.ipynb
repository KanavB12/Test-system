```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from IPython.display import Image
from IPython.core.display import HTML
import random
import timeit
```

```python
df = pd.read_csv(r'C:\Users\DELL\Desktop\ML\MissionEd\Science_Class9.csv', encoding = 'latin1')
df.head()
```

```python
df.describe()
```

```python
df = df.rename(columns = {' Question' : 'question', ' Option_A':'option_a', ' Option_B':'option_b', ' Option_C':'option_c', ' Option_D':'option_d'})
df.head()
```

```python
def test():
    start = timeit.default_timer()
    score = 0
    for i in range(10):
        index = random.randrange(len(df)-1)
        print('question ', i+1, df['question'][index])
        if(df['Images'][index] != 'None'):
            url = df['Images'][index]
            image = mpimg.imread(url)
            plt.imshow(image)
            plt.show()
        print('A)', df['option_a'][index])
        print('B)', df['option_b'][index])
        print('C)', df['option_c'][index])
        print('D)', df['option_d'][index])
        print('Enter Answer - ')
        answer = input()
        while(answer not in  ('B', 'A', 'C', 'D', 'a', 'b', 'c', 'd')):
            print('Enter valid answer')
            answer = input()
        if(answer.upper() == df['Correct_Answer'][index]):
            score += 1
    stop = timeit.default_timer()
    print('Score :', score)
    print('Time Taken : ', round((stop-start)/60, 2), 'minutes')
    labels = 'Correct', 'Incorrect'
    size = [score, 10 - score]
    plt.pie(size, labels = labels, autopct = '%1.1f%%', startangle = 90)
```

```python
test()
```
